```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../technologies.csv")
df["experience"] = df["experience"].astype(np.int8)
df = df[df["tags"].notnull()]
df.head()
```

```python
df["category"] = df["tags"].astype("category")
test = df["category"].str.get_dummies(",")
top_tags = test.sum().sort_values(ascending=False).head(10)
```

```python
top_tags.plot.bar(
    xlabel="Technology",
    ylabel="Values",
)
```
