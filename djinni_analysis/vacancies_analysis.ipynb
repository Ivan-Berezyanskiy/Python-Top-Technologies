```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../technologies.csv")
df.head()
```

```python
df["experience"] = df["experience"].astype(np.int8)
df = df[df["tags"].notnull()]
```

```python
df["tags_category"] = df["tags"].astype("category")
```

```python
tags_test = df[df["experience"] == 0]["tags_category"].str.get_dummies(",")
if not tags_test.empty:
    top_tags = tags_test.sum().sort_values(ascending=False).head(10)
    top_tags.plot.bar(
        title="0 years",
        xlabel="Technology",
        ylabel="Values",
    )
```
